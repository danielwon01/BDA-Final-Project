# 데이터 
## 데이터 병합
### list208 + list312

```python
df = pd.read_csv('./list_208_Data.csv', index_col = 0)
df.head()
```

```python
df.drop('Unnamed: 0.1',axis = 1 , inplace = True)
```

```python
df1 = df
df1
```

```python
df2 = pd.read_csv('./list_312_Data.csv', index_col = 0)
df2
```

```python
df_fix = pd.concat([df1, df2])
df_fix
```

## 데이터 전처리

```python
# size 컬럼에서 사이즈만 추출  

def size (x) :
    return str(x)[:3]

df_fix['size'] = df_fix["size"].apply(size)
df_fix
```

```python
# price 컬럼  " , " 과 " 원 " 제거 

import re

def extract_num (num_str) :
    if type(num_str) == str :
        num_str = re.sub('[^0-9]','',num_str)
    return num_str

df_fix['price'] = df_fix['price'].apply(extract_num)
df_fix
```

```python
# date , release_date 컬럼 변경 
df_fix['date'] = df_fix["date"].str.replace('/','-')
df_fix['date'] = '20' + df_fix['date']
df_fix['release_date'] = df_fix['release_date'].str.replace('/','-')
df_fix['release_date'] = '20' + df_fix['release_date']
```

```python
df_fix
```

```python
df_fix.to_csv('./data_1.csv')
```

## products 데이터 병합
